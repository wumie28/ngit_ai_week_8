# Encoding by mapping

```python
df["Gender"]=df["Gender"].map({"Male":0,"Female":1})  
```

```python
df.head(200)
```

```python
X=df.head(200)
print (X)
```

Divide features into training and testing set

```python
X_train,X_test=train_test_split(X,train_size=0.8,random_state=9)
```

```python
print (X_train)
```

```python
print(X_test)
```

# import scaler 

```python
scaler =StandardScaler()  
scaler.fit_transform(X_train)
```

```python
pipeline=Pipeline([
        ("scaler", StandardScaler()),
         ("model", KMeans(n_clusters=2,random_state=9))

])
```

```python
pipeline.fit(X_train)
```

# Custom scorer for clustering

```python
def silhouette_scorer(KMeans, X):
    labels = KMeans.fit_predict(X)
    return silhouette_score(X, labels)

```

#GridSearchCV

```python
params ={
    "model__n_clusters" : [ 2,4,6,8],
    "model__copy_x": [True,False],
    "model__algorithm" : ["lloyd","elkan"]


}


grid =GridSearchCV(estimator=pipeline, cv=3 ,param_grid=params,scoring=silhouette_scorer)
```

# fit grid on training data

```python
grid.fit(X_train)
```

```python
print(grid.best_params_)
```

# Evaluating our model by comparing with others clusters

```python
for k in range(2, 9):
    km = KMeans(n_clusters=k, random_state=42)
    labels = km.fit_predict(X)
    print(k, silhouette_score(X, labels))
```
