# Encoding by mapping

```python
df["Gender"]=df["Gender"].map({"Male":0,"Female":1})  
```

```python
X = np.array([
    [0, 19, 15, 39],
    [0, 21, 15, 81],
    [1, 20, 16,  6],
    [1, 23, 16, 77],
   [1, 31, 17,  40]

])
```

Divide features into training and testing set

```python
X_train,X_test=train_test_split(X,train_size=0.8,random_state=9)
```

# import scaler 

```python
scaler =StandardScaler()  
scaler.fit_transform(X_train)
```

```python
pipeline=Pipeline([
        ("scaler", StandardScaler()),
         ("model", KMeans(n_clusters=2,random_state=9))

])
```

```python
pipeline.fit(X_train)
```

# Custom scorer for clustering

```python
def silhouette_scorer(estimator, X):
    labels = estimator.fit_predict(X)
    return silhouette_score(X, labels)

```

GridSearchCV

```python
params ={
    "Model__n_clusters" : [ 2,4,6,8],
    "Model__copy_x": [True,False],
    "Model__algorithm" : ["lloyd","elkan"]


}


grid = GridSearchCV(estimator=pipeline, cv=3 ,param_grid=params,scoring=silhouette_scorer)
```

# fit grid on training data

```python
grid.fit(X_train)
```
